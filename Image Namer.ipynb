```python
import os

def rename_files(input_folder):
    files = os.listdir(input_folder)

    for i, old_file_name in enumerate(files):
        new_file_name = os.path.join(input_folder, f"{i:04d} (11).jpg")

        old_file_path = os.path.join(input_folder, old_file_name)
        new_file_path = os.path.join(input_folder, new_file_name)

        os.rename(old_file_path, new_file_path)
        print(f"Renamed: {old_file_path} to {new_file_path}")

if __name__ == "__main__":
    input_folder = "C:/Users/nagas/Desktop/Capstone/synthetictest"
    rename_files(input_folder)
```
