```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

dataset_path = "mixed_images"

# Get the list of classes
classes = os.listdir(dataset_path)

# Count the number of images in each class
class_counts = {cls: len(os.listdir(os.path.join(dataset_path, cls))) for cls in classes}

# Visualize class distribution with rotated x-axis labels
plt.figure(figsize=(10, 6))
sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()))
plt.title("Class Distribution in Object Classification Dataset")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
from PIL import Image
import random

dataset_path = "class_wise_train_images_fake"
# Function to display sample images from each class
def display_sample_images(dataset_path, classes, num_samples=3):
    num_classes = len(classes)
    num_cols = min(6, num_classes)  # Set the number of columns for subplots

    plt.figure(figsize=(8, 2 * (num_classes // num_cols + 1)))

    for i, cls in enumerate(classes, start=1):
        # Extract the descriptive part of the class name
        class_name = cls.split('-')[1].strip()
        
        plt.subplot((num_classes // num_cols) + 1, num_cols, i)
        plt.title(f"Class {i}: {class_name}")
        
        # Get random sample images
        sample_images = random.sample(os.listdir(os.path.join(dataset_path, cls)), num_samples)
        
        # Display the sample images
        for img_file in sample_images:
            img_path = os.path.join(dataset_path, cls, img_file)
            img = Image.open(img_path)
            plt.imshow(img)
            plt.axis("off")
            break  # Display only the first image for brevity
    
    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()

# Display sample images from each class with modified class names
display_sample_images(dataset_path, classes)
```

```python
# Function to get image sizes
def get_image_sizes(dataset_path, classes):
    image_sizes = []

    for cls in classes:
        for img_file in os.listdir(os.path.join(dataset_path, cls)):
            img_path = os.path.join(dataset_path, cls, img_file)
            img = Image.open(img_path)
            width, height = img.size
            image_sizes.append((width, height))

    return image_sizes

# Get image sizes
image_sizes = get_image_sizes(dataset_path, classes)

# Visualize image sizes
plt.figure(figsize=(10, 6))
sns.scatterplot(x=[size[0] for size in image_sizes], y=[size[1] for size in image_sizes])
plt.title("Image Sizes in Object Classification Dataset")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```

```python
import cv2
import numpy as np

# Function to get the number of color channels in each image
def get_color_channels(dataset_path, classes):
    channel_counts = []

    for cls in classes:
        for img_file in os.listdir(os.path.join(dataset_path, cls)):
            img_path = os.path.join(dataset_path, cls, img_file)
            img = cv2.imread(img_path)
            
            # Get the number of color channels
            channels = img.shape[2] if len(img.shape) == 3 else 1
            channel_counts.append(channels)

    return channel_counts

# Get color channel information
channel_counts = get_color_channels(dataset_path, classes)

# Visualize color channel distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=channel_counts)
plt.title("Color Channel Distribution in Object Classification Dataset")
plt.xlabel("Number of Color Channels")
plt.ylabel("Number of Images")
plt.show()
```
