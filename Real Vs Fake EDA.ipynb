```python
# Function to get image sizes in Real vs Fake dataset
def get_real_fake_image_sizes(dataset_path, classes):
    image_sizes = []

    for cls in classes:
        for img_file in os.listdir(os.path.join(dataset_path, cls)):
            img_path = os.path.join(dataset_path, cls, img_file)
            img = Image.open(img_path)
            width, height = img.size
            image_sizes.append((width, height))

    return image_sizes

# Get image sizes
real_fake_image_sizes = get_real_fake_image_sizes(real_fake_dataset_path, real_fake_classes)

# Visualize image sizes
plt.figure(figsize=(8, 5))
sns.scatterplot(x=[size[0] for size in real_fake_image_sizes], y=[size[1] for size in real_fake_image_sizes], hue=real_fake_classes)
plt.title("Image Sizes in Real vs Fake Dataset")
plt.xlabel("Width")
plt.ylabel("Height")
plt.legend(title="Class")
plt.show()
```
